```python
df["ParentMaritalStatus"].unique()
```

```python
group3= df.groupby(["ParentMaritalStatus"]).agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(group3)
```

```python
plt.figure(figsize=(5,5))
group3.plot(kind='bar' , color = ['pink' , 'plum' , 'blue'] ,linewidth = 0.4, edgecolor= 'black')
plt.title("Average scores considering marital status")
plt.ylabel("Aggregate average scores")
plt.axhline(group3.mean().mean() , color = 'red' , linestyle= ':' , label='Overall Mean')
plt.legend(loc="lower right")
plt.show() 
sns.heatmap(group3 , annot = True ,cmap = 'coolwarm')
plt.show()
```

```python
#From the above chart we have concluded that the marital status of the parents have no or negligible 
# impact on their child's score
```

```python
plt.title("Gender Distribution of sports practice")
chart =sns.countplot(data=df ,x='Gender', hue="PracticeSport"  )
for label in chart.containers: #for adding labels to all the containers
    chart.bar_label(label)
plt.show()
```

```python
#From the above graph we have analyzed that :
# The numbers for males and females across all three categories(regularly,sometimes,never) are quite similar, 
#suggesting that sports participation is almost equal for both genders.
#However, females have slightly higher participation numbers across all categories. 
```

```python
group4 = df.groupby(['PracticeSport']).agg({'MathScore':'mean',
                                                      'ReadingScore':'mean',
                                                      'WritingScore':'mean'}
                                                      )
print(group4)
```

```python
plt.figure(figsize=(5,5))
chart =group4.plot(kind='bar' , color = ['pink' , 'plum' , 'blue'] ,linewidth = 0.4, edgecolor= 'black')
plt.title("Average Scores based on Sports Practice")
for label in chart.containers: #for adding labels to all the containers
    chart.bar_label(label,fmt='%.1f', padding=3)
plt.show()
```

```python
#Though students who play sports regularly have slightly better aggregate scores than who practice sports somethimes
#and never
#However, there's not much impact of sports practice on student's score 
```
